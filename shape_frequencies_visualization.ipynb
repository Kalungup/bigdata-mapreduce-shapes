# Shape Frequency Visualization

This notebook loads the final MapReduce output (shape counts) and displays the result as a bar chart using Python and Matplotlib.

The counts are based on the Hadoop Streaming job results:

- circle: 6973  
- diamond: 10153  
- pentagon: 25145  
- square: 12966  
- star: 4963  
- triangle: 39800  


```python
# Shape frequency data from Hadoop MapReduce output
shapes = ["circle", "diamond", "pentagon", "square", "star", "triangle"]
counts = [6973, 10153, 25145, 12966, 4963, 39800]

list(zip(shapes, counts))
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(shapes, counts)
plt.xlabel("Shape")
plt.ylabel("Frequency")
plt.title("Shape Frequencies from Hadoop MapReduce Output")
plt.show()
```
